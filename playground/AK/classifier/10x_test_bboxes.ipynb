```python
import os
import time
import pickle
import itertools

import numpy as np
import matplotlib.pyplot as plt
from pigeon import annotate

from genEM3.data.wkwdata import WkwData,DataSource
from genEM3.util.path import getDataDir
import genEM3.data.annotation as annotation
```

```python
# Create a list of DataSource s from the bounding bboxes
bboxes=[[29839, 19074, 2358, 280, 280, 29],
        [20491, 18833, 5246, 280, 280, 29],
        [20427, 24497, 5841, 280, 280, 29],
        [28239, 23773, 5883, 280, 280, 29],
        [25342, 20650, 6907, 280, 280, 29],
        [25566, 21466, 3424, 280, 280, 29],
        [28966, 16081, 2173, 280, 280, 29],
        [19909, 16787, 1720, 280, 280, 29],
        [20550, 25007, 1172, 280, 280, 29],
        [29169, 24719, 415, 280, 280, 29],
        ]
target_binary_add = 1
target_class_add = 0.0
input_mean_add = 148.0
input_std_add = 36.0
path_add = "/tmpscratch/webknossos/Connectomics_Department/2018-11-13_scMS109_1to7199_v01_l4_06_24_fixed_mag8/color/1"

data_sources_out = []
for bbox_idx, bbox_add in enumerate(bboxes):
    data_source_out = DataSource(
        id=str(bbox_idx),
        input_path=path_add,
        input_bbox=bbox_add,
        input_mean=input_mean_add,
        input_std=input_std_add,
        target_path=path_add,
        target_bbox=bbox_add,
        target_class=target_class_add,
        target_binary=target_binary_add
    )
    data_sources_out.append(data_source_out)
```

```python
# Initialize the dataset object
input_shape = (140, 140, 1)
target_shape = (1, 1, 1)
stride = (35, 35, 1)
dataset_whole = WkwData(
    input_shape=input_shape,
    target_shape=target_shape,
    data_sources=data_sources_out,
    stride=stride,
    cache_HDD=False,
    cache_RAM=False)
```

```python
# Generate datasets for annotations
margin=35
roi_size=140
data_sources_patch = annotation.patch_source_list_from_dataset(dataset=dataset_whole, margin=margin, roi_size=roi_size)
```

```python
# Generate the datasets necessary for annotation
cache_RAM = True
cache_HDD = False
cache_HDD_root = os.path.join(getDataDir(), '.cache/')
datasets_for_annotation = dict.fromkeys(data_sources_patch)
for key, cur_source in data_sources_patch.items():
    cur_cache_HDD_dir = os.path.join(getDataDir(),'.cache'+'_'+key)
    input_shape = tuple(cur_source[0].input_bbox[3:6])
    datasets_for_annotation[key] = WkwData(input_shape=input_shape,
                            target_shape=input_shape,
                            data_sources=cur_source,
                            cache_RAM=cache_RAM,
                            cache_HDD=cache_HDD,
                            cache_HDD_root=cache_HDD_root)
```

```python
list_ranges = annotation.divide_range(total_size=len(datasets_for_annotation['large']), 
                                      chunk_size=1000)
print(list_ranges)
```

```python
#Annotate data using pigeon
annotation_fun = lambda i: annotation.display_example(i, dataset=datasets_for_annotation['large'], margin=margin, roi_size=roi_size)
annotations = []
for cur_range in list_ranges:
    print(f'Following range is {cur_range}')
    cur_a = annotate(cur_range,
                     options=['clean', 'debris', 'myelin'],
                     display_fn=annotation_fun)
    annotations.append(cur_a)
```
